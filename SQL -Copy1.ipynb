```python
import pandas
sales = pandas.read_csv("orders_csv.csv")
```

```python
#Dimension and size of the dataset
print("The dimension of the dataset is:", sales.shape)
```

```python
#Determining coumns of the dataset
print("The columns of the dataset is:", sales.columns)
```

```python
#Establishing connection with SQLite
from sqlalchemy import create_engine
engine = create_engine('sqlite://',echo=False)
```

```python
#Converting dataframe in an SQL format
sales.to_sql('salesql', con=engine)
```

```python
ans1 = engine.execute("SELECT * FROM salesql").fetchone()
print("Displaying all the fields of first record:")
for i in ans1:
    print(i)
```

```python
ans2 = engine.execute("SELECT * FROM salesql").fetchall()
print("Displaying all the fields of all records:")
for i in ans2:
    print(i)
```

```python
ans2 = engine.execute("SELECT * FROM salesql LIMIT 3").fetchall()
print("Displaying all the fields of all records:")
for i in ans2:
    print(i)
```

```python
ans2 = engine.execute("SELECT OrderID, Quantity,UnitPrice, Product_Category FROM salesql LIMIT 5").fetchall()
print("Displaying all the fields of all records:")
for i in ans2:
    print(i)
```

```python
ans5 = engine.execute("SELECT OrderID, Quantity,UnitPrice, CASE WHEN UnitPrice>200 THEN 'Premium' WHEN UnitPrice>150 THEN 'Standard' ELSE 'Cheap' END FROM salesql LIMIT 30").fetchall()
print("Using Case.. When..:")
for i in ans5:
    print(i)
```

```python
ans6 = engine.execute("SELECT * FROM salesql WHERE Quantity<=10").fetchall()
print("Using Where Clause:")
for i in ans6:
    print(i)
```

```python
ans6 = engine.execute("SELECT * FROM salesql WHERE Quantity ==100").fetchall()
print("Using Where Clause:")
for i in ans6:
    print(i)
```

```python
ans6 = engine.execute("SELECT * FROM salesql WHERE Quantity between 95 and 100").fetchall()
print("Using Where Clause:")
for i in ans6:
    print(i)
```

```python
#Distinct Address
ans7 = engine.execute("SELECT DISTINCT shipping_address FROM salesql").fetchall()
print("Using Where Clause:")
for i in ans7:
    print(i)
```
